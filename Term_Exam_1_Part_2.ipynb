```python
#Write a program that generates simple additions quizzes. The program should first display two random numbers to be added, and invite user to enter an answer. If the answer is correct, a message of congratulations should be displayed. Otherwise, a message showing the correct answer should be displayed.

import random

def math_quiz():
    num1 = random.randint(1,500)
    num2 = random.randint(1,500)
    print(f"  {num1}\n+ {num2}")
    answer = int(input("Enter sum of numbers: "))
    if answer == num1 + num2:
        print("Correct answer - Good job!")
    else:
        print(f"Incorrect! The correct answer is: {num1 + num2} ")

math_quiz()
```
